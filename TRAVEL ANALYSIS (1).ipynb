```python
def numerical(df,col):
    mean = df[col].mean()
    median = df[col].median()
    mode = df[col].mode()[0]
    min_1 = df[col].min()
    max_1 = df[col].max()
    range_1 = df[col].max() - df[col].min()
    var = df[col].var()
    std = df[col].std()
    coeffvar = (df[col].std() / df[col].mean()) * 100
    iqr = df[col].quantile(0.75) - df[col].quantile(0.25)

    print("Mean : ",mean)
    print("Median : ",median)
    print("Mode : ",mode)
    print("Minimum : ",min_1)
    print("Maximum : ",max_1)
    print("Range : ",range_1)
    print("Variance : ",var)
    print("Standard Deviation : ",std)
    print("Coefficient of Variance : ",coeffvar)
    print("IQR : ",iqr)
```

```python
def categorical(df,col):
    unique = df[col].unique()
    nunique = df[col].nunique()
    count = df[col].value_counts()

    print("Unique Values : ",unique)
    print()
    print("Number of Unique Values : ",nunique)
    print()
    print("Count : ",count)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure 'hill_station' values are properly formatted
df["hill_station"] = df["hill_station"].astype(str).str.strip().str.lower()

# Count available hill stations
num_hill_stations = df[df["hill_station"] == "yes"]["places"].nunique()
print(f"Total unique hill stations: {num_hill_stations}")

# Get place counts (without restricting to 10)
place_counts = df[df["hill_station"] == "yes"]["places"].value_counts().reset_index()
place_counts.columns = ["places", "count"]

# Use min(10, num_hill_stations) to avoid errors if less than 10 exist
place_counts = place_counts.nlargest(min(10, num_hill_stations), "count")

# Plot
plt.figure(figsize=(4, 3))
sns.barplot(data=place_counts, x="count", y="places", color="green")

plt.xlabel("Number of Mentions")
plt.ylabel("Hill Stations")
plt.title(f"Top {min(10, num_hill_stations)} Hill Stations")
plt.show()
```
