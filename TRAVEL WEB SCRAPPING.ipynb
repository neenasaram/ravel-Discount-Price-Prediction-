```python
import requests
from bs4 import BeautifulSoup
import html5lib
import re
```

```python
ur = "https://traveltriangle.com/tour-packages/india/page/1?category=friends&category=family&destination_type=india"
ur_1 = "https://traveltriangle.com/tour-packages/india/page/"
ur_2 = "?category=friends&category=family&destination_type=india"
```

```python
url_list = []
name = []
d = []
value = []
rating = []
discount = []
la = []
dur = []
place = []
stars = []
adventure = []
nature = []
hill_station = []
religious = []
water_activities = []

for i in range(1,11):
  new_url = ur_1+str(i)+ur_2

  response = requests.get(new_url)

  soup = BeautifulSoup(response.text, 'html5lib')

  url_list.append(new_url)

  for j in url_list:
    for k in soup.find_all("h3","col-md-8 pl0 fw9 m0 f16 pfc3 at_packageName"):
      name.append(k.text)

    for a in soup.find_all("h4","sfc3 m0 f24 fw9 priceVal at_newprice"):
      value.append(a.text)

    for b in soup.find_all("label","pr0 pt0"):
      rating.append(b.text)

    for p in soup.find_all("span","f12 fw4 pt2 pb2 pl8 pr8 radius20 pbc1 sfcw at_discount_label"):
      discount.append(p.text)

    for q in soup.find_all("div","_1IDFBSq tooltipBox z3 sbcw"):
      la.append(q.text)

    for e in soup.find_all("span","iblock sfc6"):
      dur.append(e.text)

    for t in soup.find_all("ul","clearfix package-cities-list at_packagecity_list"):
      shan = t.text.strip()
      place.append(shan)

    for z in soup.find_all("ul","package-incexc-list at_packageincexc-list"):
      stars.append(z.text)

    for i in soup.find_all("ul", class_="package-tags at_package_tags"):
        if "Adventure" in i.text:
            adventure.append("Yes")
        else:
            adventure.append("No")

    for i in soup.find_all("ul", class_="package-tags at_package_tags"):
        if "Nature" in i.text:
            nature.append("Yes")
        else:
            nature.append("No")

    for i in soup.find_all("ul", class_="package-tags at_package_tags"):
        if "Hill Station" in i.text:
            hill_station.append("Yes")
        else:
            hill_station.append("No")

    for i in soup.find_all("ul", class_="package-tags at_package_tags"):
        if "Religious" in i.text:
            religious.append("Yes")
        else:
            religious.append("No")

    for i in soup.find_all("ul", class_="package-tags at_package_tags"):
        if "Water Activities" in i.text:
            water_activities.append("Yes")
        else:
            water_activities.append("No")

print(url_list)
print(name)
print(len(url_list))
print(len(name))
# print(d)
# print(len(d))
print(value)
print(len(value))
print(rating)
print(len(rating))
print(discount)
print(len(discount))
print(la)
print(len(la))
print(dur)
print(len(dur))
print(place)
print(len(place))
print(stars)
print(len(stars))
print("adventure: ",adventure)
print(len(adventure))
print("nature: ",nature)
print(len(nature))
print("Hill Station: ",hill_station)
print(len(hill_station))
print("religious: ",religious)
print(len(religious))
print("water_activities: ",water_activities)
print(len(water_activities))
```
